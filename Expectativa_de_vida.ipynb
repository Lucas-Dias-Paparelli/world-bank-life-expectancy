```python
## Tratando os dados faltantes

# Selecioando apenas as colunas númericas
numeric_columns = df.select_dtypes(include=['number'])

# Preenchendo os valores ausentes nas colunas numericas com as médias das respectivas colunas

df = df.fillna(numeric_columns.mean()) # Passei direto df.fillna, pois eu sei que apenas as colunas numericas contém valores ausentes
```

***
Essa linha de código trata os valores faltantes no DataFrame, preenchendo-os com a média aritmética de cada coluna. A escolha de preencher os valores faltantes com a média aritmética foi feita para manter a integridade dos dados e preservar a estrutura geral do conjunto de dados, minimizando a perda de informações.

Ao preencher os valores faltantes com a média de cada coluna, estamos mantendo uma abordagem simples e direta, que é adequada especialmente quando não é possível estimar os valores ausentes de forma mais precisa e quando os dados apresentam uma distribuição relativamente homogênea. Isso permite que continuemos a análise dos dados sem a necessidade de remover linhas ou colunas significativas, evitando assim uma perda substancial de informações.

```python
## Modelagem Preditiva de Expectativa de Vida usando Árvore de Decisão

# Importando as bibliotecas

import numpy as np # Para fazer a SEED e sempre que rodarmos o algoritmo ter o mesmo resultado
from sklearn.model_selection import train_test_split # Para dividir os dados
from sklearn.tree import DecisionTreeRegressor # Algoritmo da árvore de decisão
from sklearn.metrics import mean_squared_error, r2_score # Métricas para medir o qual bom o algoritmo esta sendo

# Separar as features (x) e a variável alvo (y)

x = df[['Prevelance of Undernourishment', 'CO2',
        'Health Expenditure %', 'Education Expenditure %',
        'Unemployment', 'Corruption', 'Sanitation', 'Injuries',
        'Communicable', 'NonCommunicable']]
y = df['Life Expectancy World Bank']

# Separando os dados

SEED = 158020

np.random.seed(SEED)

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)

# Inicialização do modelo

model = DecisionTreeRegressor()

# Treinando o Modelo

model.fit(x_train, y_train)

# Previsões

y_pred = model.predict(x_test)

# Avaliação

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Printando a Avaliação

print('Erro Quadrático Médio (MSE):', mse.round(3))
print()
print("Coeficiente de Determinação (R²):", r2.round(3))

```

***
**Conclusão:**

Com base na aplicação do modelo de árvore de decisão, observamos resultados promissores na previsão da expectativa de vida com base nas variáveis selecionadas. O Erro Quadrático Médio (MSE) de aproximadamente 2.912 anos indica que, em média, as previsões do modelo estão desviando cerca de 3 anos da expectativa de vida real. Além disso, o Coeficiente de Determinação (R²) de 0.956 sugere que o modelo é capaz de explicar aproximadamente 96,6% da variabilidade na expectativa de vida com base nas variáveis fornecidas. Esses resultados demonstram que o modelo de árvore de decisão apresenta um desempenho robusto na previsão da expectativa de vida, fornecendo uma ferramenta valiosa para análises futuras e tomada de decisões relacionadas à saúde pública e políticas de bem-estar.
